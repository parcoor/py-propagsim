```python
import pandas as pd
```

```python
df = pd.read_csv('../doc/prop_gender.csv')
df
```

```python
p_gender = {'h': {}, 'f': {}}
for i, row in df.iterrows():
    pre = 0 if i == 0 else 15 + 5 * (i - 1)
    for g in ['h', 'f']:
        p_gender[g][pre] = row[g]
        
print(p_gender)
    
```

```python
f80 = 6 * .65
h80 = 6 * .35

p_gender['f'][75] = p_gender['f'][75] - f80
p_gender['h'][75] = p_gender['h'][75] - h80
p_gender['f'][80] = f80
p_gender['h'][80] = h80
```

```python
for g in ['h', 'f']:
    p_gender[g] = {k: round(v, 1) for k, v in p_gender[g].items()}
```

```python
import json

with open('../data/p_gender_pop.json', 'w') as f:
    json.dump(p_gender, f)
```
